```python
#Q2 using two-ptr method
str = input("Enter string: ")
n = len(str)
for i in range(n//2):
  if str[i] != str[n-i-1]:
    print("Not a Pallindrome")
    break
else:
    print("Is a Pallindrome")

```

```python
#Q3
nums = [int(x) for x in input("Enter integers: ").split()]
x = int(input("Enter requiered difference: "))
for i in range(len(nums)):
  for j in range(i+1 , len(nums)):
    if abs(nums[j] - nums[i]) == x:
      print(nums[i] , nums[j])

```

```python
#Q4
nums = [int(x) for x in input("Enter integers: ").split()]
n = len(nums)

for i in range(n):
  if min(nums) < nums[i] < max(nums):
    break
nums[i], nums[n-1] = nums[n-1], nums[i]

pvt = nums[n-1]
i = 0
j = 0
k = n-1

while i <= k:
  if nums[i] < pvt:
    nums[i], nums[j] = nums[j], nums[i]
    i += 1
    j += 1
  elif nums[i] == pvt:
    i += 1
  else:
    nums[i], nums[k] = nums[k], nums[i]
    k -= 1

print(nums)

```
